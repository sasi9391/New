```python
a2=np.array([2,4,7])
const=3
print(a2&const)#0010 & 0011 = 0010 = 2,0100 & 0011 = 0000 = 0,0111 & 0011 = 0011 =3-----> both the conditions true
print(a2|const) # atleast one condition true
print(a2^const)# same conditions is there it will gives true
print(a2<<const)#[16 32 56]
print(a2>>const)#[0 0 0]
```

```python
print((a<2) & (a<5))#[ True False False False False False False False]
print((a==2) | (a==5))#[False  True False False False False False  True]
print(~(a==5))#[ True False  True  True  True  True  True False]
print(a[~(a==5)])#[1 2 3 4 6 7 8]
print(a[(a<2) & (a<5)])#[1]
print(a[(a==2) | (a==5)])#[2 5]
```

```python
df['Age'].isnull().sum()
```

```python
age_mod=df['Age'].mode()[0]
df['Age']=df['Age'].fillna(age_mod)
```

```python
imdb_med=df['IMDb'].median()
df['IMDb']=df['IMDb'].fillna(imdb_med)
```

```python
rotten_mod=df['Rotten Tomatoes'].mode()[0]
df['Rotten Tomatoes']=df['Rotten Tomatoes'].fillna(rotten_mod)
```

```python
directors_mod=df['Directors'].mode()[0]
df['Directors']=df['Directors'].fillna(directors_mod)
```

```python
runtime_med=df['Runtime'].median()
df['Runtime']=df['Runtime'].fillna(runtime_med)
```

```python
genres_md=df['Genres'].mode()[0]
df['Genres']=df['Genres'].fillna(genres_md)
```

```python
country_md=df['Country'].mode()[0]
df['Country']=df['Country'].fillna(country_md)
```

```python
lang_md=df['Language'].mode()[0]
df['Language']=df['Language'].fillna(lang_md)
```

```python
age_mod=df['Age'].mode()[0]
imdb_med=df['IMDb'].median()
rotten_mod=df['Rotten Tomatoes'].mode()[0]
directors_mod=df['Directors'].mode()[0]
genres_md=df['Genres'].mode()[0]
country_md=df['Country'].mode()[0]
lang_md=df['Language'].mode()[0]
runtime_med=df['Runtime'].median()
for col,val in zip(['Age','IMDb','Rotten Tomatoes','Directors','Genres','Country','Language','Runtime'],
                   [age_mod,imdb_med,rotten_mod,directors_mod,genres_md,country_md,lang_md,runtime_med]):
    df[col]=df[col].fillna(val)
```

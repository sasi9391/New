```python
import numpy as np
```

```python
l=([1,2,3,4,5,6,7])
a1=np.arange(5)
a2=np.arange(1,5)
a3=np.arange(1,5,2)
print(a1)
print(a2)
print(a3)
```

```python
#Linspace
b1=np.linspace(1,11,3,retstep=True)
print(b1)
b2=np.linspace(1,10,4,retstep=True,endpoint=True)
print(b2)
```

```python
#Zeros
print(np.zeros((2,)))
print(np.zeros((2,3)))
print(np.zeros((2,3,4)))
```

```python
#ones
print(np.ones((2,)))
print(np.ones((2,3,4)))
print(np.ones((2,3,4)))

#outer=2,inner=3,val=4
```

```python
#Full
print(np.full((2,3),7))
print(np.full((2,5),(5)))
#outer=2,inner values=3, perticular value to print=7
```

```python
#eye
#identity matrix will come
print(np.eye(3))
print(np.eye(3,4))
```

```python
import numpy as np
```

```python
#car speed  (k/d)
speed=[20,30,40,35,50]
np.array([speed])
m=np.mean(speed)
print("mean-->",m)
v=np.var(speed)
print("var-->",v)
s=np.std(speed)
print("std-->",s)#fluctuations
cv=(s/m)*100
print("cv-->",cv)
```

```python
d=np.array([20,500])
print(d)
print(d.dtype)#int having only 8 bytes 
print(d.itemsize)#64/8
print(d.size)
print(d.nbytes)# 8*2
print(d.shape)# how many elements present inthe list that len will come with coma(,)  and tuple formate
```

```python
d=np.array([20,500,55],dtype=np.int32)
print(d)
print(d.dtype)
print(d.itemsize)# individual size 32/8=4
print(d.size)
print(d.nbytes)#complete array 4*3=12
```

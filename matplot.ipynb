```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.scatter(x,y)
plt.title('sales growth')
plt.xlabel('year')
plt.ylabel('sales')
plt.show()
plt.plot(x,y,color='red',linestyle='--',marker='*',markersize=10,markerfacecolor='blue')
```

```python
plt.plot(x,[5,6,7,8,9],color='green',linestyle='-',marker='*',markersize=10,markerfacecolor='yellow')
plt.legend(['sales','profit'])
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.bar(x,y)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.plot(x,y)
plt.grid(True)
```

```python
plt.savefig('sales growth.png')
```

```python
# object oriented interface
fig,ax=plt.subplots()
ax.plot(x,y)
ax.set_title('sales growth')
ax.set_xlabel('year')
ax.set_ylabel('sales')
plt.show()
```

```python
age=pd.Series([63,200,1,32,33,31,32,32,25,29,26,66,52]
)
age    
```

```python
Q1=np.percentile(age,25)
Q1  #29
```

```python
Q3=np.percentile(age,75)
Q3 
```

```python
IQR=Q3-Q1
IQR
```

```python
p=IQR*1.5
p
```

```python
lower_bond=Q1-p
lower_bond
```

```python
upper_bond=Q3+p
upper_bond
```

```python
sns.boxplot(x=age)
```

```python
for i in age:
    if i<=lower_bond or i>upper_bond:
        print(i)
```

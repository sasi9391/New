```python
import numpy as np
```

```python
a1=np.array([20,100,15,40,30,100,5,6])
print(np.max(a1))
print(np.min(a1))
print(np.sum(a1))
print(np.mean(a1))#40+30/2=39.5
print(np.sum(a1)/a1.size)
print(np.sum(a1)/len(a1))
print(np.median(a1))# it will arrange in ascending order ofter it will give mean value
print((a1))
print(np.median(a1,overwrite_input=True))
print(a1)
```

```python
#percentile
a1=np.array([20,100,15,40,30,5,6])
print(np.median(a1,overwrite_input=True))
print(a1)
# which num 100% of the data is present
print(np.percentile(a1,100))
print(np.percentile(a1,50))#belo 100% of the data is present
print(np.percentile(a1,0))# below the 50% data is present
print(np.percentile(a1,2))# 2 is the percentage
print(a1)
```

# 1. Avarage Dissolve

```python
df['Dissolved - Min']=df['Dissolved - Min'].str.strip()
df['Dissolved - Min']=pd.to_numeric(df['Dissolved - Min'],errors='coerce')
df=df.dropna(subset=['Dissolved - Min'])
df['Dissolved - Min']=df['Dissolved - Min'].astype(float)
df['Avg_dissolve']=df[ ['Dissolved - Max','Dissolved - Min'] ].mean(axis=1)

df['Avg_dissolve']
```

# 2. avg Conductivity

```python
df['Avg_conductivity']=df['Conductivity (¬µmho/cm) - Min'].str.strip()
df['Conductivity (¬µmho/cm) - Min']=pd.to_numeric(df['Conductivity (¬µmho/cm) - Min'],errors='coerce')
df=df.dropna(subset=['Conductivity (¬µmho/cm) - Min'])
df['Conductivity (¬µmho/cm) - Min']=df['Conductivity (¬µmho/cm) - Min'].astype(float)
df['Avg_conductivity']=df[ ['Conductivity (¬µmho/cm) - Min','Conductivity (¬µmho/cm) - Max'] ].mean(axis=1)

df['Avg_conductivity']
```

# 3. Avg temparature

```python
df['Avg_temparature']=df[ ['Temperature (C) - Min','Temperature (C) - Max'] ].mean(axis=1)
```

# 4. Avg BOD

```python
df['Avg_BOD']=df['BOD (mg/L) - Min'].str.strip()
df['BOD (mg/L) - Min']=pd.to_numeric(df['BOD (mg/L) - Min'],errors='coerce')
df=df.dropna(subset=['BOD (mg/L) - Min'])
df['BOD (mg/L) - Min']=df['BOD (mg/L) - Min'].astype(float)
df['Avg_BOD']=df[ ['BOD (mg/L) - Min','BOD (mg/L) - Max'] ].mean(axis=1)

df['Avg_BOD']
```

# 5. Avg  NitrateN

```python
df['Avg_Nitrate']=df['NitrateN (mg/L) - Min'].str.strip()
df['NitrateN (mg/L) - Min']=pd.to_numeric(df['NitrateN (mg/L) - Min'],errors='coerce')
df=df.dropna(subset=['NitrateN (mg/L) - Min'])
df['NitrateN (mg/L) - Min']=df['NitrateN (mg/L) - Min'].astype(float)
df['Avg_Nitrate']=df[ ['NitrateN (mg/L) - Min','NitrateN (mg/L) - Max'] ].mean(axis=1)

df['Avg_Nitrate']
```

## Conclusion

* The analysis of the Indian Water Quality dataset highlights the urgent need for sustainable water management and treatment strategies across the country. The data indicates that many regions face challenges with contamination from parameters such as fluoride, arsenic, salinity, and hardness, which pose risks to both human health and agriculture. While certain states show relatively safe water quality, others require immediate intervention through purification technologies, awareness programs, and stricter policy enforcement.

* This study reinforces that improving water infrastructure, promoting community-level treatment solutions, and implementing continuous monitoring are essential steps to ensure access to safe drinking water. By integrating scientific data, government initiatives, and public participation, India can make significant progress towards achieving clean and safe water for all.
